```python

def consolidator_country():
        
        consol = Consolidator(report_date = report_date,
                #weeks_data = weeks_data,
                income = income,
                expense = expense,
                savings = savings,
                repayments = repayments,
                given  = given,
                #employee = employee,
                taken = taken,
                lgiven = lgiven)

        data = [
                consol.income_normalize(),
                consol.expense_normalize(),
                consol.savings_normalize(),
                consol.repayment_normalize(),
                consol.given_normalize(),
                #consol.employee_normalize()
                ]
        consolidated = pd.concat(data);print(consolidated.shape)
        consolidated.head()
        consolidated.transaction_type = np.where((consolidated.file=='savings'),'cash',
                                        consolidated.transaction_type)
        consolidated.file = np.where((consolidated.file=='savings') & 
        (consolidated.reason_name=='Deposited'), 'savings_deposit', consolidated.file)
        consolidated.file = np.where((consolidated.file=='savings') & 
        (consolidated.reason_name=='Withdrawn'), 'savings_withdrawal', consolidated.file)

        # load data
        data = [consol.loantaken_normalize(),consol.loangiven_normalize()]
        loan_schedule = pd.concat(data)
        loan_schedule = loan_schedule.drop(["repaid_amount","remaining_amount"],1)
        loan_schedule = loan_schedule.rename(columns = {"original_amount":"amount"})
        consolidated = consolidated.append(loan_schedule, sort=False)

        # further cleaning
        consolidated["transaction_date"] = pd.DatetimeIndex(consolidated["transaction_date"])
        consolidated["week"] = ["Week "+ str(i.isocalendar()[1]) for i in consolidated["transaction_date"]]
        consolidated = consolidated[['firm_id', 'country_of_residence','data_id','amount','currency',
                                'transaction_date','account','reason_name','description', 'file',"week","name","email","phone_number"]]
        consolidated.head()

        # finalizing teh consolidation 

        file_lst = ['expense', 'savings_deposit', 'loans_taken_repayment','loans_given']

        consolidated['amount1'] = np.where(consolidated.file.isin(file_lst),consolidated.amount*-1,
        consolidated.amount)

        consolidated = consolidated[['firm_id',"name","email","phone_number", 'country_of_residence','data_id','amount','currency',
                                'transaction_date','account','reason_name','description', 'file',"week","amount1"]]
        return consolidated
```
